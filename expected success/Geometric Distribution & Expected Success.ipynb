### Geometric distribution
In probability theory and statistics, the geometric distribution is either one of two discrete probability distributions:

The probability distribution of the number X of Bernoulli trials needed to get one success, supported on the set { 1, 2, 3, ... }
The probability distribution of the number Y = X − 1 of failures before the first success, supported on the set { 0, 1, 2, 3, ... }

To be simple:

第一次成功所需要的伯努利实验的次数 $X$， 给定单次伯努利实验的概率 $p$

```python
from scipy import stats
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
plt.style.use('bmh')
```

```python
trials = list(range(30))
prob = [stats.geom(0.1).pmf(i) for i in trials]
```

```python
figsize(15,4)
plt.bar(trials, prob)
plt.xlabel('Number of trials')
plt.ylabel('Probability')
plt.title('Geometric distribution', fontsize = 15)
plt.show()
```

在一个成功概率为10%的伯努利实验中，疯狂实验，你仍然是有可能，30次才得到第一次成功，但是概率很小

```python
cdfs = [stats.geom(0.1).cdf(i) for i in trials]

figsize(15,4)
plt.bar(trials, cdfs)
plt.xlabel('Number of trials')
plt.ylabel('Probability')
plt.title('Cumulative distribution', fontsize = 15)
plt.show()
```

但我们可以看到，30次之内成功的概率基本就是1了，肯定成功了

```python
橙色战舰的概率 = 5 / 1000
```

```python
trials = [1,5,10,100,200,300,400,500,827,1000]
prob = [stats.geom(橙色战舰的概率).pmf(i) for i in trials]
```

```python
figsize(15,4)
plt.bar(trials, prob, width = 10)
plt.xlabel('Number of trials')
plt.ylabel('Probability')
plt.title('Probability Mass Function', fontsize = 15)
plt.show()
```

```python
figsize(15,4)
cdfs = [stats.geom(橙色战舰的概率).cdf(i) for i in trials]
plt.bar(trials, cdfs, width = 10)
plt.xlabel('Number of trials')
plt.ylabel('Probability')
plt.title('Cumulative distribution', fontsize = 15)
plt.show()
```

事实证明，827次实验基本肯定能抽到一艘橙色战舰了，这个哥们儿，真的小概率了
